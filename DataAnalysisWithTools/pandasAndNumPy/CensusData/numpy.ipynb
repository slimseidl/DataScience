```python
import numpy as np
```

```python
#  create random array of 100million values between 1 and 0 
a = np.random.random(int(1e8))
```

```python
# Compute mean of array with standard python
# took 5 seconds
import time
start=time.time()
sum(a) / len(a)
print(time.time() - start, 'seconds')
```

```python
# compute mean with numpy
# takes under .1 seconds
start = time.time()
np.mean(a)
print(time.time() - start, 'seconds')
```

# Measures of Spread

```python
import pandas as pd 
```

```python
df_census = pd.read_csv('census_income_data.csv')
```

```python
# quantiles 
# set your own quantiles

df_census[["capital-gain", "capital-loss"]].quantile(q=[0.25, 0.5, 0.75, 0.9, 0.95])
```

```python
df_census["workclass"].value_counts()
```

```python
# most of the data is 0, we will only look at the 50%, 90%, and 95% quantiles
# focus only on "age", "capital-gain", "capital-loss", and "hours-per-week"

df_census.groupby(by="workclass")[["age", "capital-gain", "capital-loss", "hours-per-week"]].quantile(q=[0.5, 0.9, 0.95])
```

```python
df_census.groupby("workclass")[["age","capital-gain","capital-loss","hours-per-week"]].mean()
```

```python
df_census.groupby("workclass")[["age","capital-gain","capital-loss","hours-per-week"]].std()
```

```python
df_census.groupby("workclass")[["age","capital-gain","capital-loss","hours-per-week"]].var()
```
